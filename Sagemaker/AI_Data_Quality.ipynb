```python
import pandas as pd

# Training Dataset
data = {
    "missing_values": [0, 2, 5, 0, 8, 1, 4, 0],
    "duplicate_rows": [0, 0, 2, 1, 5, 0, 3, 0],
    "invalid_salary": [0, 1, 2, 0, 3, 0, 1, 0],
    "status": [0, 1, 1, 0, 1, 0, 1, 0]
}

df = pd.DataFrame(data)

print(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

X = df[["missing_values","duplicate_rows","invalid_salary"]]
y = df["status"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42
)

model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

model.fit(X_train, y_train)

pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test,pred))
```

```python
# Example: Predict a new dataset

new_data = [[1, 0, 0]]
# Format:
# [missing_values, duplicate_rows, invalid_salary]

prediction = model.predict(new_data)
probability = model.predict_proba(new_data)

if prediction[0] == 0:
    print("Prediction: Healthy Dataset")
else:
    print("Prediction: High Risk Dataset")

print("Confidence:", round(max(probability[0]) * 100, 2), "%")
```

```python
import pandas as pd

csv_file = "employee.csv"

df_csv = pd.read_csv(csv_file)

rows = len(df_csv)
missing_values = df_csv.isnull().sum().sum()
duplicate_rows = df_csv.duplicated().sum()

invalid_salary = 0
if "salary" in df_csv.columns:
    invalid_salary = pd.to_numeric(
        df_csv["salary"],
        errors="coerce"
    ).isnull().sum()

features = [[missing_values, duplicate_rows, invalid_salary]]

prediction = model.predict(features)
probability = model.predict_proba(features)

print("="*50)
print("      AI DATA QUALITY ANALYSIS REPORT")
print("="*50)

print(f"Dataset           : {csv_file}")
print(f"Total Rows        : {rows}")
print(f"Missing Values    : {missing_values}")
print(f"Duplicate Rows    : {duplicate_rows}")
print(f"Invalid Salary    : {invalid_salary}")

print("-"*50)

if prediction[0] == 0:
    print("Prediction        : HEALTHY DATASET")
    print("Recommendation    : Proceed with Glue ETL")
else:
    print("Prediction        : HIGH RISK DATASET")
    print("Recommendation    : Move to Failed Bucket")

print(f"Confidence        : {round(max(probability[0])*100,2)} %")

print("="*50)
```
